```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

```python
#get unique count of value in the screen name column
players_count = len(purchase_data_df["SN"].unique())

#add count to a summary data frame
player_count_summary_df = pd.DataFrame({"Total Count": [players_count]})
player_count_summary_df
```

```python
#format float values to inlcude $
pd.options.display.float_format = '${:.2f}'.format

#unique item count
item_count = len(purchase_data_df["Item Name"].unique())

#average price per item
price_average = purchase_data_df["Price"].mean()
round(price_average,2)

#Number of purchases
purchases_count = len(purchase_data_df["Purchase ID"])
purchases_count

#Total Revenue calculation
total_revenue = round(purchase_data_df["Price"].sum(),2)
total_revenue

purchasing_analysis_summary_df = pd.DataFrame({"Number of Unique Items": [item_count], 
                                              "Average Price": [price_average], 
                                              "Number of Purchases": [purchases_count], 
                                              "Total Revenue": [total_revenue]})
purchasing_analysis_summary_df
#d = dict(selector="th",
 #   props=[('text-align', 'center')])

#add values to a summary data frame
#purchasing_analysis_summary_df #= pd.DataFrame({"Number of Unique Items": [item_count], "Average Price": [price_average], "Number of Purchases": [purchases_count], "Total Revenue": [total_revenue]})
#purchasing_analysis_summary_df.style.set_properties(**{'width':'15em', 'text-align':'center'})\
 #       .set_table_styles([d])
```

```python
#new dataframe to include data with no player duplicates
player_no_dupes_df = purchase_data_df.drop_duplicates(subset=['SN'])

#variables for gender counts
only_male_count = len(player_no_dupes_df.loc[player_no_dupes_df["Gender"] == "Male", :])
only_female_count = len(player_no_dupes_df.loc[player_no_dupes_df["Gender"] == "Female", :])
only_other_count = len(player_no_dupes_df.loc[player_no_dupes_df["Gender"] == "Other / Non-Disclosed", :])

#calculations for gender averages
male_average_players = (only_male_count/players_count) * 100
female_average_players = (only_female_count/players_count) * 100
other_average_players = (only_other_count/players_count) * 100


#format float values to inlcude $
#pd.options.display.float_format = '{:.2f}%'.format

#add summary data frame
gender_demo_summary_df = pd.DataFrame({"": ['Male', 'Female', 'Other/Non-Disclosed'],"Total Count": [only_male_count, only_female_count, only_other_count],
                                    "Percentage of Players": [male_average_players, female_average_players, other_average_players]})
#remove the index from the dataframe
gender_demo = gender_demo_summary_df.set_index('')
gender_demo
```
